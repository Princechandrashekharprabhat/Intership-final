```python
import matplotlib.pyplot as plt 
x=[3,8,5,15]
x1=[4,9,6,16]
y=[5,13,7,21]
y1=[4,12,6,16]
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph')
plt.scatter(x,y,marker='*',s=120,label='water')
plt.scatter(x1,y1,marker='^',s=120,label='soil')
plt.plot(x,y)
plt.plot(x1,y1)
plt.bar(x,y)
plt.bar(x1,y1)
plt.legend(loc=4)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
players=['virat','dhoni','pathan','sachin']
runs=[280,234,120,380]
plt.legend(loc=4)
plt.grid()
ply1=[]
run1=[]
for i in range(len(players)):
    if runs[i] > 200:
        ply1.append(players[i])
        run1.append(runs[i])
plt.xlabel('ply1')
plt.xlabel('runs1')
plt.title('200 and more run in a match')
plt.scatter(ply1,run1,marker='*',s=120,label='runs')
plt.plot(ply1,run1)
plt.bar(ply1,run1)
plt.legend(loc=4)
plt.grid()
plt.show()

```
