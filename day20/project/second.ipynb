```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
import seaborn as sb
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
import time

 
data_set="Advertising.csv"
df=pd.read_csv(data_set)
labels=df['Sales']
feature=df[['TV','Newspaper' ,'Radio']]
print(labels)
print(feature)
fig,ax =plt.subplots(3,figsize=(6,6))
sb.pairplot(df,x_vars=['TV','Newspaper' ,'Radio'],y_vars=['Sales'],kind='scatter')
plt.show()
train_X,test_x,train_Y,test_y=train_test_split(feature, labels,test_size=0.2)
model=LinearRegression()
trained_model=model.fit(train_X,train_Y)
predicted_output=trained_model.predict(test_x)
print("predicted_output",predicted_output)
time.sleep(2)
print("actual output",test_y)
plt.xlabel('actual sale')
plt.ylabel('predicted sale')
plt.title('marketting salescompany')
plt.scatter(test_y,predicted_output)
plt.plot([min(test_y),max(test_y)],[min(predicted_output),max(predicted_output)],color='red',linestyle='--')
plt.show()



```
